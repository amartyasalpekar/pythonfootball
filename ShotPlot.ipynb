```python
pitch = Pitch(pitch_color = 'green', line_color = 'white')
fig, ax = pitch.draw(figsize=(10, 7))

pitchLengthX = 120
pitchWidthY = 80

for i, shot in shots.iterrows():
    x = shot['x']
    y = shot['y']
    goal = shot['outcome_name'] == 'Goal'
    team_name = shot['team_name']
    circleSize = 2
    
    if (team_name == team1):
        if goal:
            shotCircle = plt.Circle((x,y),circleSize, color = "white")
            plt.text(x+1, y-2, shot['player_name'])
        else:
            shotCircle = plt.Circle((x,y),circleSize, color = "white")
            shotCircle.set_alpha(0.2)
    
    else:
        if goal:
            shotCircle = plt.Circle((pitchLengthX - x,pitchWidthY - y),circleSize, color = "blue")
            plt.text(pitchLengthX-x+1, pitchWidthY-y-2, shot['player_name'])
        else:
            shotCircle = plt.Circle((pitchLengthX-x,pitchWidthY-y),circleSize, color = "blue")
            shotCircle.set_alpha(0.2)
    ax.add_patch(shotCircle)
    
fig.suptitle("Real Madrid (White) and Barcelona (Blue) shots", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()   
```

```python
shots1 = df.loc[(df['type_name'] == "Shot") | (df['type_name'] == "Own Goal For")]

# Filter out penalty shots
non_penalty_shots = shots1.loc[shots1['period'] <= 4]
```

```python
import pandas as pd
pitch = Pitch(pitch_color = 'green', line_color = 'white')
fig, ax = pitch.draw(figsize=(10, 7))

pitchLengthX = 120
pitchWidthY = 80

for i, shot in non_penalty_shots.iterrows():
    x = shot['x']
    y = shot['y']
    goal = shot['outcome_name'] == 'Goal'or shot['type_name'] == "Own Goal For"
    team_name = shot['team_name']
    circleSize = 2
    
    player_name = shot['player_name'] if pd.notna(shot['player_name']) else 'Own Goal'

    
    if (team_name == team1):
        if goal:
            shotCircle = plt.Circle((x,y),circleSize, color = "white")
            plt.text(x+1, y-2, shot['player_name'])
            
        else:
            shotCircle = plt.Circle((x,y),circleSize, color = "white")
            shotCircle.set_alpha(0.2)
    
    else:
        if goal:
            shotCircle = plt.Circle((pitchLengthX - x,pitchWidthY - y),circleSize, color = "blue")
            plt.text(pitchLengthX-x+1, pitchWidthY-y-2, shot['player_name'])
        else:
            shotCircle = plt.Circle((pitchLengthX-x,pitchWidthY-y),circleSize, color = "blue")
            shotCircle.set_alpha(0.2)
    ax.add_patch(shotCircle)
    
fig.suptitle("Real Madrid (White) and Barcelona (Red) shots", fontsize = 12)
plt.show()   
```

```python
#create pitch
pitch1 = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#query
mask_sui = (df.type_name == 'Shot') & (df.team_name == team1)
#finding rows in the df and keeping only necessary columns
df_sui = df.loc[mask_sui, ['x', 'y', 'outcome_name', "player_name"]]

#plot them - if shot ended with Goal - alpha 1 and add name
#for Switzerland
for i, row in df_sui.iterrows():
    if row["outcome_name"] == 'Goal':
    #make circle
       pitch1.scatter(row.x, row.y, alpha = 1, s = 500, color = "white", ax=ax['pitch'])
       pitch1.annotate(row["player_name"], (row.x + 1, row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch1.scatter(row.x, row.y, alpha = 0.2, s = 500, color = "white", ax=ax['pitch'])

mask_esp = (df.type_name == 'Shot') & (df.team_name == team2)
df_esp = df.loc[mask_esp, ['x', 'y', 'outcome_name', "player_name"]]

#for Sweden we need to revert coordinates
for i, row in df_esp.iterrows():
    if row["outcome_name"] == 'Goal':
       pitch1.scatter(120 - row.x, 80 - row.y, alpha = 1, s = 500, color = "blue", ax=ax['pitch'])
       pitch.annotate(row["player_name"], (120 - row.x + 1, 80 - row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch1.scatter(120 - row.x, 80 - row.y, alpha = 0.2, s = 500, color = "red", ax=ax['pitch'])

fig.suptitle("Real Madrid (White) and Barcelona (Red) shots", fontsize = 30)
plt.show()
```
