```python
from matplotlib.patches import FancyArrow

#drawing pitch
pitch = Pitch(line_color = "white", pitch_color='green')
fig, ax = pitch.draw(figsize=(10, 7))

for i,thepass in messi_pass.iterrows():
    #if pass made by Leo Messi
    x=thepass['x']
    y=thepass['y']
        #plot circle
    passCircle=plt.Circle((x,y),2,color="white")
    passCircle.set_alpha(.2)
    ax.add_patch(passCircle)
    dx=thepass['end_x']-x
    dy=thepass['end_y']-y
    #plot arrow
    passArrow = FancyArrow(x, y, dx, dy, width=0.3, head_width=1, head_length=1, color="white")
    ax.add_patch(passArrow)
    
    if thepass['outcome_name'] == 'Incomplete':
        # Plot an 'x' at the end of the arrow for incomplete passes
        ax.plot(thepass['end_x'], thepass['end_y'], marker='x', color='red', markersize=8, markeredgewidth=2)


ax.set_title("Leo Messi passes against Barcelona", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```

```python
mask_barca = (df.type_name == 'Pass') & (df.team_name == "Barcelona") & (df.sub_type_name != "Throw-in")
df_passes = df.loc[mask_barca, ['x', 'y', 'end_x', 'end_y', 'player_name']]
#get the list of all players who made a pass
names = df_passes['player_name'].unique()
```

```python
pitch = Pitch(pitch_color = 'green', line_color = 'white', pad_top = 20)
fig, axs = pitch.grid(ncols = 4, nrows = 3, grid_height=0.85, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0.04, endnote_space=0.01)

for name, ax in zip(names, axs['pitch'].flat[:len(names)]):
    ax.text(60, -10, name,
            ha='center', va='center', fontsize=14)
    #take only passes by this player
    player_df = df_passes.loc[df_passes["player_name"] == name]
    #scatter
    pitch.scatter(player_df.x, player_df.y, alpha = 0.2, s = 50, color = "white", ax=ax)
    #plot arrow
    pitch.arrows(player_df.x, player_df.y,
            player_df.end_x, player_df.end_y, color = "white", ax=ax, width=1)

#We have more than enough pitches - remove them
for ax in axs['pitch'][-1, 16 - len(names):]:
    ax.remove()

#Another way to set title using mplsoccer
axs['title'].text(0.5, 0.5, 'Barcelona (3) passes against Real Madrid (2)', ha='center', va='center', fontsize=30)
plt.show()
```
